```python
df= pd.read_csv("test.csv")
```

```python
df.isnull().any(axis=0)
```

```python
df["Artist Reputation"] = df["Artist Reputation"].replace(np.NaN ,df["Artist Reputation"].median())
df["Height"] = df["Height"].replace(np.NaN ,df["Height"].median())
df["Weight"] = df["Weight"].replace(np.NaN ,df["Weight"].median())
df["Width"] = df["Width"].replace(np.NaN ,df["Width"].median())

df['Material'].fillna(df['Material'].mode()[0], inplace=True)
df['Transport'].fillna(df['Transport'].mode()[0], inplace=True)
df['Remote Location'].fillna(df['Remote Location'].mode()[0], inplace=True)
```

```python
df.isnull().any(axis=0)
```

```python
fdata = [df]
for dataset in fdata:
    # Mapping Material
    dataset['Material'] = dataset['Material'].map( {'Brass': 0, 'Clay': 1, 'Aluminium': 2, 'Wood': 3, 'Marble': 4, 'Bronze': 5, 'Stone': 6} ).astype(int)
    # Mapping International
    dataset['International'] = dataset['International'].map( {'No': 0, 'Yes': 1} ).astype(int)
    # Mapping Express Shipment
    dataset['Express Shipment'] = dataset['Express Shipment'].map( {'No': 0, 'Yes': 1} ).astype(int)
    # Mapping Fragile
    dataset['Fragile'] = dataset['Fragile'].map( {'No': 0, 'Yes': 1} ).astype(int)
    # Mapping Coustomer Information
    dataset['Customer Information'] = dataset['Customer Information'].map( {'Working Class': 0, 'Wealthy': 1} ).astype(int)
    # Mapping Installation Included
    dataset['Installation Included'] = dataset['Installation Included'].map( {'No': 0, 'Yes': 1} ).astype(int)
    # Mapping Installation Included
    dataset['Remote Location'] = dataset['Remote Location'].map( {'No': 0, 'Yes': 1} ).astype(int)
    # Mapping Coustomer Information
    dataset['Transport'] = dataset['Transport'].map( {'Airways': 0, 'Roadways': 1, 'Waterways': 2} ).astype(int)
```
